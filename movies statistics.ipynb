### Using 'groupby()' to aggregate

```python
runtime_genre = df[['Genres','Runtime']].groupby('Genres').mean()
print(runtime_genre.head())
```

```python
imdb_language = df[['Language','IMDb']].groupby('Language').mean().dropna()
print(imdb_language)
```

```python
def get_group_mean(categorical,numerical):
    group_df = df[[categorical, numerical]].groupby(categorical).mean().dropna()
    print(group_df.head())
```

```python
get_group_mean('Directors','IMDb')
```
